# Colors of markers

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.mpl_style', 'default')

CSV_IN = '../data/polish-police+astro.csv'
```

```python
df = pd.read_csv(CSV_IN, parse_dates=['Data'], dayfirst=True)
# To jako tip:
# df['Mercury retro'] = np.where(df['Mercury speed'] < 0,'red','blue'); df[['Mercury retro']][:5]
```

```python
plt.figure(1,figsize=[16,8], dpi=144)
cm = plt.cm.get_cmap('spring')
z = df['Mercury speed']
plt.scatter(df['Sun'], df['Wypadki drogowe'], c=z, vmin=-1.3, vmax=2.2,  s=30, cmap=cm, alpha=.8)
plt.xlim(0,360)
plt.xticks([0,30,60,90,120,150,180,210,240,270,300,330,360])
plt.xlabel('Dugo ekliptyczna Soca')
plt.ylabel('Ilo wypadkw drogowych')
plt.title('Ilo wypadkw drogowych w Polsce i dugo ekliptyczna Soca\n (Merkury w retro zaznaczono kolorem fioletowym)\n')
plt.colorbar(label='Szybko Merkurego')
plt.figtext(0.5,0.03,s='rdo danych: www.policja.pl/pol/form/1,dok.html, Dane za okres 01.12.2008-25.03.2016, Opracowa: Tomasz Stpor',alpha=.8, ha='center')
# plt.show()
```
