```python
%matplotlib inline
import numpy as np
import pandas as pd
pd.set_option('display.mpl_style', 'default')
import matplotlib.pyplot as plt

CSV = '../data/polish-police.csv'
```

```python
df1 = pd.read_csv(CSV, parse_dates=['Data'], dayfirst=True)
# df2 = df1.drop_duplicates('Data').set_index('Data')
df2 = df1.drop_duplicates('Data')
df2.plot(figsize=(16,4))
```

```python
plt.figure(1,figsize=[16,8], dpi=144)
plt.title('Wypadki drogowe w Polsce\n',)
plt.plot_date(df1['Data'],df1['Wypadki drogowe'], alpha=.5, markersize=10, color='yellow')
plt.plot_date(df1['Data'],df1['Ranni w wypadkach'], alpha=.4, markersize=5, color='red', marker="^")
plt.plot_date(df1['Data'],df1['Zabici w wypadkach'], alpha=.5, color='violet')
plt.legend(['Wypadki drogowe', 'Ranni w wypadkach', 'Zabici w wypadkach'])
plt.ylabel('Ilo zdarze')
plt.figtext(0.5,0.07,s='Lata',alpha=.8, ha='center')
plt.figtext(0.5,0.03,s='rdo danych: www.policja.pl/pol/form/1,dok.html, Opracowa: @TomaszStapor',alpha=.8, ha='center')
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, figsize=[16,6])

ax1.plot_date(df1['Data'],df1['Wypadki drogowe'], alpha=.3, markersize=5, color='red')
ax1.plot_date(df1['Data'],df1['Zatrzymani nietrzewi kierujcy'], alpha=.3, markersize=5)
ax1.set_title('Wypadki drogowe i zatrzymani nietrzewi kierujcy\n')
ax1.set_ylim(0,1000)

ax2.scatter(df1['Wypadki drogowe'], df1['Zatrzymani nietrzewi kierujcy'], alpha=.3)
ax2.set_title('Korelacja\n')
ax2.set_xlabel('Ilo wypadkw drogowych')
ax2.set_ylabel('Ilo nietrzewych kierujcych')
ax2.set_ylim(0,1000)
ax2.set_xlim(0,200)
```
