# Polish Police polar plots

```python
%matplotlib inline
import numpy as np
import pandas as pd
pd.set_option('display.mpl_style', 'default')
import matplotlib.pyplot as plt
import datetime

CSV = '../data/polish-police+astro.csv'

# df1 = pd.read_csv(CSV, parse_dates=['Data'], dayfirst=True, index_col='Data')
df1 = pd.read_csv(CSV, parse_dates=['Data'], dayfirst=True)
```

## Get only 2009-2015 years

```python
df2 = df1.loc[(df1['Data'].dt.year > 2008) & (df1['Data'].dt.year < 2016)]
```

## Sun eliptic longitude

```python
fig = plt.figure(figsize=[14,8], dpi=144)
fig.suptitle("Ilości wypadków i zatrzymań nietrzeźwych w latach 2009-2015\nwg położnia Słońca na ekliptyce\n", fontsize=16)
theta = df2['Sun'] * np.pi / 180
ax = plt.subplot(121, projection='polar')
ax.set_theta_offset(1*np.pi)
ax.set_thetagrids([0,30,60,90,120,150,180,210,240,270,300,330])
ax.set_rlim(200)
ax.set_title('Wypadki drogowe\n')
ax.plot(theta, df2['Wypadki drogowe'], 'bo', color='r', linewidth=3, alpha=.3)
ax2 = plt.subplot(122, projection='polar')
ax2.set_theta_offset(1*np.pi)
ax2.set_thetagrids([0,30,60,90,120,150,180,210,240,270,300,330])
ax2.set_rlim(1000)
ax2.set_title('Zatrzymani nietrzeźwi kierujący\n')
ax2.plot(theta, df2['Zatrzymani nietrzeźwi kierujący'], 'bo', color='b', linewidth=3, alpha=.3)
plt.figtext(0.5,0.1,s='Dane za okres 01.01.2009-31.12.2015, Opracował: Tomasz Stąpor, Źródło danych: www.policja.pl/pol/form/1,dok.html', ha='center')
plt.show()
```
