# Retrogradation of Mercury and road accidents in Poland

```python
%matplotlib inline
import numpy as np
import pandas as pd
pd.set_option('display.mpl_style', 'default')
import matplotlib.pyplot as plt
import datetime

CSV = '../data/polish-police+astro.csv'

# df1 = pd.read_csv(CSV, parse_dates=['Data'], dayfirst=True, index_col='Data')
df1 = pd.read_csv(CSV, parse_dates=['Data'], dayfirst=True)
```

## First

```python
cm = plt.cm.get_cmap('copper')
fig = plt.figure(figsize=[16,8], dpi=144)
fig.suptitle("Merkury w retro a wypadki", fontsize=24)

def printGraph(df, title):
    N = 50
    size = df['Wypadki drogowe'] / N
    x = df['Wypadki drogowe']
    y = df['Wypadki drogowe']
    colors = df['Mercury speed']
    area = (np.pi * size*1.2)**2  # 0 to 15 point radiuses
    plt.scatter(df['Data'].dt.dayofyear, y, s=area, c=colors, alpha=0.7,cmap=cm)
    plt.colorbar(label='Szybko Merkurego')
    plt.xlim(0,365)
#     plt.xticks([0,30,60,90,120,150,180,210,240,270,300,330])
    plt.ylim(0,200)
    plt.title(title)

ax1 = plt.subplot(231)
printGraph(df1.loc[(df1['Data'].dt.year == 2009)], '2009')
ax2 = plt.subplot(232)
printGraph(df1.loc[(df1['Data'].dt.year == 2010)], '2010')
ax1 = plt.subplot(233)
printGraph(df1.loc[(df1['Data'].dt.year == 2011)], '2011')
ax2 = plt.subplot(234)
printGraph(df1.loc[(df1['Data'].dt.year == 2012)], '2012')
ax1 = plt.subplot(235)
printGraph(df1.loc[(df1['Data'].dt.year == 2013)], '2013')
ax2 = plt.subplot(236)
printGraph(df1.loc[(df1['Data'].dt.year == 2014)], '2014')

plt.show()
```

```python
import matplotlib as mpl
cm = plt.cm.get_cmap('cool', 4 )
fig = plt.figure(figsize=[16,10], dpi=144)
fig.suptitle("Merkury w retro a liczba wypadkw", fontsize=24)

def printGraph(df, title, alpha=0.7):
    N = 50
    size = df['Wypadki drogowe'] / N
    x = df['Wypadki drogowe']
    y = df['Wypadki drogowe']
    colors = df['Mercury speed']
    area = (np.pi * size*1.2)**2  # 0 to 15 point radiuses
    plt.scatter(df['Data'].dt.dayofyear, y, s=area, c=colors, alpha=alpha,cmap=cm)
#     vmin=-1, vmax=2
    plt.xlim(0,365)
    plt.xticks([0,30,60,90,120,150,180,210,240,270,300,330], ('sty', 'lut', 'mar', 'kwi', 'maj', 'cze', 'lip', 'sie', 'wrz', 'pa', 'lis', 'gru'))
    plt.ylim(0,250)
    plt.title(title)
    
plt.subplot2grid((6,2), (0,0), rowspan=2, colspan=2)
printGraph(df1.loc[(df1['Data'].dt.year >= 2010) & (df1['Data'].dt.year <= 2015)], 'Razem za lata 2010-2015', 0.7)
plt.colorbar(label='Szybko Merkurego')

plt.subplot2grid((6,2), (2,0))
printGraph(df1.loc[(df1['Data'].dt.year == 2010)], '2010')

plt.subplot2grid((6,2), (3,0))
printGraph(df1.loc[(df1['Data'].dt.year == 2011)], '2011')

plt.subplot2grid((6,2), (4,0))
printGraph(df1.loc[(df1['Data'].dt.year == 2012)], '2012')

plt.subplot2grid((6,2), (2,1))
printGraph(df1.loc[(df1['Data'].dt.year == 2013)], '2013')

plt.subplot2grid((6,2), (3,1))
printGraph(df1.loc[(df1['Data'].dt.year == 2014)], '2014')

plt.subplot2grid((6,2), (4,1))
printGraph(df1.loc[(df1['Data'].dt.year == 2015)], '2015')
plt.figtext(0.5,0.21,s='Wielko kropki zaley od liczby wypadkw, kolor od szybkoci Merkurego', ha='center', fontsize=12)
plt.figtext(0.5,0.18,s='Opracowa: Tomasz Stpor, rdo danych: www.policja.pl/pol/form/1,dok.html', ha='center', alpha=.7)

plt.show()
```
