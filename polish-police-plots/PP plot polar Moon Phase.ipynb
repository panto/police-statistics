# Moon Phases

```python
%matplotlib inline
import numpy as np
import pandas as pd
pd.set_option('display.mpl_style', 'default')
import matplotlib.pyplot as plt
import datetime

CSV = 'polish-police+astro.csv'

# df1 = pd.read_csv(CSV, parse_dates=['Data'], dayfirst=True, index_col='Data')
df1 = pd.read_csv(CSV, parse_dates=['Data'], dayfirst=True)
```

## Moon phases

```python
theta = df1['MoonSun distance'] * np.pi / 180

fig = plt.figure(figsize=[16,8], dpi=144)
fig.suptitle("Faza Ksiyca", fontsize=24)

ax = plt.subplot(121, projection='polar')
ax.set_theta_offset(0.5*np.pi)
ax.set_thetagrids([0,30,60,90,120,150,180,210,240,270,300,330],
                 labels=['0', '30', '60', '90', '120', '150', '180', '-150', '-120', '-90', '-60', '-30'])
ax.set_rlim(200)
ax.set_title('Wypadki drogowe\n')
ax.set_rlabel_position(0)
ax.plot(theta, df1['Wypadki drogowe'], 'bo', color='r', linewidth=3, alpha=.2)

ax2 = plt.subplot(122, projection='polar')
ax2.set_theta_offset(0.5*np.pi)
ax2.set_thetagrids([0,30,60,90,120,150,180,210,240,270,300,330],
                 labels=['0', '30', '60', '90', '120', '150', '180', '-150', '-120', '-90', '-60', '-30'])
ax2.set_rlim(1000)
ax2.set_title('Zatrzymani nietrzewi kierujcy\n')
ax2.set_rlabel_position(0)
ax2.plot(theta, df1['Zatrzymani nietrzewi kierujcy'], 'bo', color='b', linewidth=3, alpha=.2)

plt.figtext(0.5,0.1,s='Dane za okres 01.12.2008-25.03.2016', ha='center', fontsize=12)
plt.figtext(0.5,0.05,s='Opracowa: Tomasz Stpor, rdo danych: www.policja.pl/pol/form/1,dok.html', ha='center', alpha=.8)
plt.show()
```
